```python
import requests
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

def search(was, wo, seit, seite):
    """search for jobs. params can be found here: https://jobsuche.api.bund.dev/"""
    params = {
        'angebotsart': '1',
        'page': seite,
        'pav': 'false',
        'size': '250', # das maximum sind 250 
        'umkreis': '25',
        'was': was,
        'wo': wo,
        'veroeffentlichtseit' : seit
    }
    headers = {
        'User-Agent': 'Jobsuche/2.9.2 (de.arbeitsagentur.jobboerse; build:1077; iOS 15.1.0) Alamofire/5.4.4',
        'Host': 'rest.arbeitsagentur.de',
        'X-API-Key': 'jobboerse-jobsuche',
        'Connection': 'keep-alive',
    }
    response = requests.get(
        'https://rest.arbeitsagentur.de/jobboerse/jobsuche-service/pc/v4/app/jobs',
        headers=headers, params=params, verify=False
    )
    return response.json()   



result = search("", "obertraubling","1","250")
print(type(result))
print(result.keys())     
print(result)


```

```python
result = search("", "obertraubling","10","1")
for nummer, job in enumerate(result.get("stellenangebote", [])):
    print(job.get("titel"), job.get("arbeitgeber"), nummer+1)  
```
